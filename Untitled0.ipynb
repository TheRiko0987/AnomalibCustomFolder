```python
# Import the datamodule
from anomalib.data import Folder

import torch

from anomalib.deploy import ExportType, TorchInferencer

from torchvision.transforms.v2 import Compose, RandomAdjustSharpness, RandomHorizontalFlip, Resize, Normalize

from anomalib.models import Patchcore

train_transform = Compose(
    [
        RandomHorizontalFlip(p=0.5),
        Resize((256, 256)),
        Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ],
)

eval_transform = Compose(
    [
        Resize((256, 256)),
        Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ],
)


data_root = "/content/pack"

# Create the datamodule
datamodule = Folder(
    name="pack",
    root=data_root,
    normal_dir="train",
    abnormal_dir="test",
    task="classification",
    train_transform=train_transform,
    eval_transform=eval_transform
)

# Setup the datamodule
datamodule.setup()
```

```python
# Access the datasets
train_dataset = datamodule.train_data
val_dataset = datamodule.val_data
test_dataset = datamodule.test_data

# Access the dataloaders
train_dataloader = datamodule.train_dataloader()
val_dataloader = datamodule.val_dataloader()
test_dataloader = datamodule.test_dataloader()
```

```python
i, train_data = next(enumerate(datamodule.train_dataloader()))
print(train_data.keys())

i, val_data = next(enumerate(datamodule.val_dataloader()))
print(val_data.keys())

i, test_data = next(enumerate(datamodule.test_dataloader()))
print(test_data.keys())
```

```python
print(train_data["image"].shape)
```

```python
# Import the model and engine
from anomalib.engine import Engine
from anomalib import TaskType

# Create the model and engine
model = Patchcore()
engine = Engine()

# Train a Patchcore model on the given datamodule
engine.fit(model=model, datamodule=datamodule)
```

```python
# load best model from checkpoint before evaluating
test_results = engine.test(
    model=model,
    datamodule=datamodule,
    ckpt_path=engine.trainer.checkpoint_callback.best_model_path,
)
print(test_results)
```
